```python
# Read a diabetes dataset
df = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv')
```

# CLEANING STEP:

CLEANING STEP, TASK 1:  In column "weight," replace values "?" with "0"

```python
df['weight'] = df['weight'].replace('?', '0')
```

```python
df['weight'].unique()
```

```python
df['weight'].value_counts()
```

CLEANING STEP, TASK 2:  Replace all NaN's with 0

```python
df.fillna(0, inplace=True)
```

```python
# Check dataset for NaN's:
df.isnull().sum()
```

Observe: No more NaN's in dataset, as NaN's have been replaced by '0'

# SOME DESCRIPTIVE STATISTICS AND VISUALIZATIONS:

```python
df.describe()
```

Observe above:  Out of the 50 columns, the numerical columns are

**encounter_id	patient_nbr	admission_type_id	discharge_disposition_id	admission_source_id	time_in_hospital	num_lab_procedures	num_procedures	num_medications	number_outpatient	number_emergency	number_inpatient	number_diagnoses**

Below will focus on the column "**TIME_IN_HOSPITAL**" which--according to the Data Dictionary--means "Integer number of days between admission and discharge."

```python
df['time_in_hospital'].describe()
```

Observe above:  For all 101766 patients in the dataset, the mean number of days between admission and discharge is about 4.39 days with a maximum number of days as 14.

```python
df['time_in_hospital'].value_counts()
```

```python
# "time_in_hospital" is 1 through 14 (no zeroes), so 14 (not 15) values,
# so use 14 for number of bins in histogram command:
df['time_in_hospital'].hist(bins=14)
plt.title('OVERALL Histogram of "time_in_hospital"')
plt.xlabel('Days')
plt.ylabel('Number of Patients')
plt.show()
```

Observe that the "**TIME_IN_HOSPITAL**" distribution is not normal but right-skewed.  The mode is the third rectangle (from left-to-right), corresponding to more than 17500 patients.  Compared to the dataset total of 101766 rows, this is about 18% of patients.  Further, spending 3 days or less in the hospital for diabetes is about 48% of all patients.

```python
df['race'].value_counts()
```

```python
sns.barplot(x=df['race'].value_counts().index, y=df['race'].value_counts())
plt.title('Count of Patients by Race')
plt.xlabel('Race')
plt.ylabel('Number of Patients')
plt.show()
```

```python
df['race'].describe()
```

Observe above:  Column "race" has six values:  "Caucasian," "AfricanAmerican," "Hispanic," "Asian," "Other," and "Unknown/Invalid" (denoted by '?').  The top two most frequent values are "Caucasian" and "AfricanAmerican."  There are 76099 Caucasian patients (74.8% of the total) and 19210 AfricanAmerican patients (18.9% of the total).  For the remaining, there are 6457 patients of other races or of those whose race is unknown.

```python
df_Caucasian = df[df['race'] == 'Caucasian']
df_AfrAmer = df[df['race'] == 'AfricanAmerican']
df_Hispanic = df[df['race'] == 'Hispanic']
df_Asian = df[df['race'] == 'Asian']
df_Other = df[df['race'] == 'Other']
df_Unknown = df[df['race'] == '?']
```

```python
# Caucasian Race "time_in_hospital"
df_Caucasian['time_in_hospital'].hist(bins=14, color='red')
plt.ylim(ymin=0, ymax=14000)
plt.title('Caucasian Race Histogram of "time_in_hospital"')
plt.xlabel('Days')
plt.ylabel('Number of Patients')
plt.show()
```

```python
# AfricanAmerican Race "time_in_hospital"
df_AfrAmer['time_in_hospital'].hist(bins=14, color='blue')
plt.ylim(ymin=0, ymax=14000)
plt.title('AfricanAmerican Race Histogram of "time_in_hospital"')
plt.xlabel('Days')
plt.ylabel('Number of Patients')
plt.show()
```

```python
# Hispanic Race "time_in_hospital"
df_Hispanic['time_in_hospital'].hist(bins=14, color='green')
plt.ylim(ymin=0, ymax=14000)
plt.title('Hispanic Race Histogram of "time_in_hospital"')
plt.xlabel('Days')
plt.ylabel('Number of Patients')
plt.show()
```

**Overplotting Histograms of Caucasian, AfricanAmerican, and Hispanic "time_in_hospital":**

```python
df_Caucasian['time_in_hospital'].plot(kind='hist', bins=14, color='red')
df_AfrAmer['time_in_hospital'].plot(kind='hist', bins=14, color='blue')
df_Hispanic['time_in_hospital'].plot(kind='hist', bins=14, color='green')
```

```python
# Hispanic Race "time_in_hospital" ("SCALED TO ymax of 500")
df_Hispanic['time_in_hospital'].hist(bins=14, color='green')
plt.ylim(ymin=0, ymax=500)
plt.title('Hispanic Race Histogram of "time_in_hospital"')
plt.xlabel('Days')
plt.ylabel('Number of Patients')
plt.show()
```

```python
# Asian Race "time_in_hospital" ("SCALED TO ymax of 500")
df_Asian['time_in_hospital'].hist(bins=14, color='orange')
plt.ylim(ymin=0, ymax=500)
plt.title('Asian Race Histogram of "time_in_hospital"')
plt.xlabel('Days')
plt.ylabel('Number of Patients')
plt.show()
```
